```python
import numpy as np
import pandas as pd
```

```python
def searchColumn(word, cols):
    return [col for col in cols if word in col]
```

```python
data_path = "../data/rawdata.csv"
rawdata = pd.read_csv(data_path, header=0)
```

```python
rawdata
```

```python
# 正規化

# MSM
UGRD, VGRD, TMP, RH, PRMSL, PRES, APCP
U,V以外を標準化

# 平塚塔(measured)
'measured_UGRD', 'measured_VGRD'
そのまま

# 気象庁
rain tmp sun wind_velは標準化
風向はそのまま（クラスだから）

# ブイ
buoy_tmp
```

```python
msm_col = searchColumn("msm", rawdata.columns)
searchColumn("SS", rawdata.columns)
```

```python
# 標準化するparam
params = ["TMP", "RH", "PRMSL", "PRES", "APCP", "tmp", "rain", "sun", "wind"]

param_cols = []
for param in params:
    cols = searchColumn(param, rawdata.columns)
    param_cols.extend(cols)
```

```python
rawdata[param_cols].describe()
```

```python
_std_data = rawdata[param_cols].apply(lambda x: (x - np.mean(x)) / np.std(x))
```

```python
_std_data.describe()
```

```python
_std_data
```

```python
std_data = pd.DataFrame()
for col in rawdata.columns:
    try:
        std_data[col] = _std_data[col]
    except KeyError:
        std_data[col] = rawdata[col]
```

```python
std_data.to_csv("../data/std_data.csv", index=False)
```
