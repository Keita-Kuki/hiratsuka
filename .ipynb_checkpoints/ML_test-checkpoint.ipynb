# road data

```python
avg_10min_data_path = "../data/hiratsuka/avg_10min"
year = 2017

data_path = "{0}/{1}".format(avg_10min_data_path, year)

vel_avg = []
for month in range(7, 8 + 1):
    for day in range(1, 31 + 1):
        file_name = "{0}{1:02d}{2:02d}.csv".format(year, month, day)
        file_dath = "{0}/{1}".format(data_path, file_name)

        data = np.genfromtxt(file_dath, delimiter=",", dtype='float')
        vel_avg.extend(data[:, 1])

data_hiratsuka_10min = pd.DataFrame(columns=["time", "vel"])
data_hiratsuka_10min["time"] = pd.date_range('2017-07-01 00:00:00', periods=8928 ,freq='10min')
data_hiratsuka_10min["vel"] = vel_avg
```

```python
avg_1h_data_path = "../data/hiratsuka/avg_1h"
year = 2017

data_path = "{0}/{1}".format(avg_1h_data_path, year)

vel_avg = []
for month in range(7, 8 + 1):
    for day in range(1, 31 + 1):
        file_name = "{0}{1:02d}{2:02d}.csv".format(year, month, day)
        file_dath = "{0}/{1}".format(data_path, file_name)

        data = np.genfromtxt(file_dath, delimiter=",", dtype='float')
        vel_avg.extend(data[:, 1])

data_hiratsuka_1h = pd.DataFrame(columns=["time", "vel"])
data_hiratsuka_1h["time"] = pd.date_range('2017-07-01 00:00:00', periods=1488 ,freq='H')
data_hiratsuka_1h["vel"] = vel_avg
```

```python
def get_vel_data(year, lon, lat):
    csv_dir_path="../data/csv"
    file_path="{0}/{1}/{2}".format(csv_dir_path, lon, lat)

    ugrd_file_name="{0}/{1}_{2}_UGRD_{3}.csv".format(file_path, lon, lat, year)
    vgrd_file_name="{0}/{1}_{2}_VGRD_{3}.csv".format(file_path, lon, lat, year)

    ugrd_data = np.genfromtxt(ugrd_file_name, delimiter=",", dtype='float')
    vgrd_data = np.genfromtxt(vgrd_file_name, delimiter=",", dtype='float')

    return (ugrd_data[:,1]**2 + vgrd_data[:, 1]**2)**0.5

year = 2017
lons=["139.2500", "139.3125", "139.3750", "139.4375"]
lats=["35.20", "35.25", "35.30", "35.35", "35.40"]

data_msm = pd.DataFrame(columns=["time"])
data_msm["time"] = pd.date_range('2017-07-01 9:00:00', periods=1488, freq='H')

i = 0
for lon in lons:
    for lat in lats:
        data_msm["vel_{0}".format(i)] = get_vel_data(year, lon, lat)
        i = i + 1
```

# 平塚１時間平均を予測

```python
test_NN = Sequential()
test_NN.add(Dense(10, input_dim = 20))
test_NN.add(Activation('sigmoid'))
test_NN.add(Dense(10))
test_NN.add(Activation('sigmoid'))
test_NN.add(Dense(1))
test_NN.summary()
test_NN.compile(optimizer = 'adam', loss = 'mse')
```

```python
start = datetime(2017,7,1,9,0,0)
end = start + timedelta(days=62) + timedelta(hours=-9)
# print(start, end)

data_y = data_hiratsuka_1h.query('time >= \"{0}\" & time < \"{1}\"'.format(start, end))
data_x = data_msm.query('time >= \"{0}\" & time < \"{1}\"'.format(start, end))

input = data_x[['vel_0', 'vel_1', 'vel_2', 'vel_3', 'vel_4', 'vel_5', 'vel_6', 'vel_7', 'vel_8', 'vel_9', 'vel_10', 'vel_11', 'vel_12', 'vel_13', 'vel_14', 'vel_15', 'vel_16', 'vel_17', 'vel_18', 'vel_19']]
input = np.array(input)
label = np.array(data_y["vel"])
test_NN.fit(input, label, epochs = 100)
```

```python
p = test_NN.predict(input)
predict = pd.DataFrame(columns=["time", "vel"])
predict["time"] = pd.date_range('2017-07-01 09:00:00', periods=1479 ,freq='H')
predict["vel"] = p
predict
```

```python
def plot_wind(start, end, predict):
    
    fig = plt.figure(figsize=(18, 12), dpi=80)
    ax1 = fig.add_subplot(2,1,1)

    data_1 = data_hiratsuka_1h.query('time >= \"{0}\" & time < \"{1}\"'.format(start, end))
    data_2 = data_msm.query('time >= \"{0}\" & time < \"{1}\"'.format(start, end))
    data_3 = predict.query('time >= \"{0}\" & time < \"{1}\"'.format(start, end))

    ax1.plot(data_1["time"], data_1["vel"], label = "Hiratsuka")
    ax1.plot(data_2["time"], data_2["vel_0"], label = "MSM")
    ax1.plot(data_3["time"], data_3["vel"], label = "predict")

    plt.legend()
    plt.show()
#     png_name = "images/wind_vel/wind_vel_{0}_{1}.png".format(start.strftime("%Y%m%d"), end.strftime("%Y%m%d"))
#     plt.savefig(png_name)
```

```python
delta = 7
start = datetime(2017,7,1,0,0,0)
end = start + timedelta(days=delta)

while end < datetime(2017,9,1,0,0,0):
    plot_wind(start, end)
    start = start + timedelta(days=delta)
    end = start + timedelta(days=delta)
```
